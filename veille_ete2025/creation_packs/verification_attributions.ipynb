# Verification des attributions des articles entre veilleurs

```python
import pandas as pd 
import numpy as np 

import warnings
```

```python
warnings.filterwarnings("ignore")

# Pandas options
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 200)
```

```python
base = pd.read_csv("attribution_articles_veilleurs_08072025.csv", index_col=0)
premium = pd.read_csv("attribution_articles_veilleurs_pack_premium_09072025.csv", index_col=0)
```

# Sur le pack de base

```python
print(f"le pack de base contient {base.nom_fichier.nunique()} articles répartis entre {base.veilleur_id.nunique()} veilleurs")
```

```python
# Nombre de répetitions de chaque article
base.groupby("nom_fichier")["veilleur_id"].nunique()
```

```python
base.groupby("nom_fichier")["veilleur_id"].value_counts().reset_index()
```

```python
# Nombre d'articles différents par veilleurs
base.groupby("veilleur_id")["nom_fichier"].value_counts().reset_index()
```

## Sur le pack premium

```python
print(f"le pack premium contient {premium.nom_fichier.nunique()} articles répartis entre {premium.veilleur_id.nunique()} veilleurs")
```

```python
# Nombre de répetitions de chaque article
premium.groupby("nom_fichier")["veilleur_id"].nunique()
```

```python
premium.groupby("nom_fichier")["veilleur_id"].value_counts().reset_index()
```

```python
# Nombre d'articles différents par veilleurs
premium.groupby("veilleur_id")["nom_fichier"].value_counts().reset_index()
```

## Verification de l'absence d'articles en commun dans le pack de base et le pack premium

```python
liste_base = base.nom_fichier.unique().tolist()
liste_premium = premium.nom_fichier.unique().tolist()
```

```python
for article in liste_base:
    if article in liste_premium:
        print(f"L'article {article} est dans les 2 listes")
    else:
        print(f"Article {article}: 😎")
```

```python
for article in liste_premium:
    if article in liste_base:
        print(f"L'article {article} est dans les 2 listes")
    else:
        print(f"Article {article}: 😎")
```
